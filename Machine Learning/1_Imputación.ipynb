## Valores perdidos e imputación

Creamos valores

```python
import random
```

```python
energydata.shape # dimensiones
```

```python
# creamos valores aleatorios con semilla
random.seed(6)
lista_aleatoria = random.sample(range(19735), 1000) # genera 1000 números aleatorios de 0 a 19734
```

```python
energydata_missing = energydata.copy()
energydata_missing.loc[lista_aleatoria,"Appliances"] = np.nan # np.nan asigna valores NA
```

```python
## Se han generado 1000 valores perdidos en la variable Appliances para utilizar los métodos de imputación
energydata_missing.describe().T
```

Imputación por medidas de tendencia central

```python
# imputación por la media
energydata_missing_media  = energydata_missing.fillna(energydata_missing['Appliances'].mean()).copy()
```

```python
# imputación por la mediana
energydata_missing_median = energydata_missing.fillna(energydata_missing['Appliances'].median()).copy()
```

**Imputación por algoritmo KNN**

```python
from sklearn.impute import KNNImputer

energydata_missing_knn = energydata_missing.copy()
# Construimos el modelo KNNImputer con 3 vecinos y pesos uniformes
imputer = KNNImputer(n_neighbors=3, weights="uniform")

# Ajustamos el modelo e imputamos los missing values
imputer.fit(energydata_missing_knn[['Appliances']])

# Añadimos los valores imputados
energydata_missing_knn['Appliances'] = imputer.transform(energydata_missing_knn[['Appliances']]).ravel()
```

```python
energydata[['Appliances']].describe().T # valores iniciales
```

```python
energydata_missing_media[['Appliances']].describe().T # valores con imputación con la media
```

```python
energydata_missing_median[['Appliances']].describe().T # valores con imputación con la mediana
```

```python
energydata_missing_knn[['Appliances']].describe().T # valores con imputación con KNN
```

**Imputación por Regresión Lineal**

```python
energydata_missing_lm = energydata_missing.copy()
energydata_missing_lm_complete = energydata_missing_lm[~energydata_missing_lm.Appliances.isna()] # todos los registros que no tenga datos faltantes en "Appliances"
```

```python
print(energydata_missing_lm.shape)
print(energydata_missing_lm_complete.shape)
```

```python
## En un modelo de regresión debemos escoger SOLAMENTE la variables numéricas como "predictoras"
## Por lo que si queremos predecir la variable "Appliances" no puede ser parte de las predictoras
## Y la variable 'date' se descarta porque no es una variable numérica.
parameters = [x for x in energydata_missing_lm.columns if x not in ['date','Appliances']]
```

```python
parameters
```

```python
from sklearn import linear_model

# creamos el modelo
model = linear_model.LinearRegression()
model.fit(X = energydata_missing_lm_complete[parameters], y = energydata_missing_lm_complete['Appliances']) # entrenamos con los 18735 observaciones (sin los valores faltantes, ya que eso queremos imputarlos)
```

```python
energydata_missing_lm['Appliances'].isnull() # toda los valores de Appliances, pero viendo si es nulo o no
```

```python
variable_imputada = model.predict(energydata_missing_lm[parameters])[energydata_missing_lm['Appliances'].isnull()]
# energydata_missing_lm['Appliances'].isnull() seleccionando los valores NA de Appliances
# energydata_missing_lm[parameters]: seleccionando las variables
# variable_imputada
```

```python
energydata_missing_lm['Appliances_imp'] = energydata_missing_lm['Appliances']
# se crea otra variable con las mismas observaciones de Appliances
# el objetivo es poder reemplazarla con los valores imputados en el código posterior
energydata_missing_lm[['Appliances','Appliances_imp']].describe().T
```

```python
energydata_missing_lm.loc[energydata_missing_lm['Appliances'].isnull(), 'Appliances_imp'] = variable_imputada
# Cuando la data sea nula, entonces se tomará los valores de "variable_imputada"
#energydata_missing_lm.describe().T
```

```python
energydata_missing_lm[['Appliances','Appliances_imp']].describe().T
```

**Imputación con Random Forest**

```python
energydata_missing_rf = energydata_missing.copy()
energydata_missing_rf_complete = energydata_missing_rf[~energydata_missing_rf.Appliances.isna()] # todos los registros que no tenga datos faltantes en "Appliances"
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
```

```python
# La ejecución demora un poco porque el modelo construye N arboles
model_rf = RandomForestRegressor(n_estimators = 100, random_state = 9)
# número de arboles = n_estimators
model_rf.fit(X = energydata_missing_rf_complete[parameters], y = energydata_missing_rf_complete['Appliances'])
```

```python
variable_imputada_rf = model_rf.predict(energydata_missing_rf[parameters])[energydata_missing_rf['Appliances'].isnull()]
# energydata_missing_rf['Appliances'].isnull() seleccionando los valores NA de Appliances
# energydata_missing_rf[parameters]: seleccionando las variables
```

```python
energydata_missing_rf['Appliances_imp'] = energydata_missing_rf['Appliances']
# se crea otra variable con las mismas observaciones de Appliances
# el objetivo es poder reemplazarla con los valores imputados en el código posterior
```

```python
energydata_missing_rf.loc[energydata_missing_rf['Appliances'].isnull(), 'Appliances_imp'] = variable_imputada_rf
# lo reemplazamos con los valores imputados
```
